```python
# import Libraries
import pandas as pd
import numpy as np
from pathlib import Path 
```

```python
#import Raw csv to pandas dataframe
DATA_DIR = Path(r'C:\Users\Admin\OneDrive\Documents\levelz-project\raw_csv')
df = pd.read_csv(DATA_DIR / 'new_retail_data.csv')
```

```python
print(df.shape)
df.head()
```

```python
# checking for missing vlaues
print(df.isnull().sum())
```

```python
# Handling missing values
# Drop missing values
df = df.dropna(subset=['Transaction_ID', 'Customer_ID'])
df.shape
```

```python
# fill categorical fields with placeholders (unknown)

categorical_cols = ['Gender','Customer_Segment','Shipping_Method', 'Payment_Method','Order_Status']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```
