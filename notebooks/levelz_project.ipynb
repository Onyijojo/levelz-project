```python
# import Libraries
import pandas as pd
import numpy as np
from pathlib import Path 
```

```python
#import Raw csv to pandas dataframe
DATA_DIR = Path(r'C:\Users\Admin\OneDrive\Documents\levelz-project\raw_csv')
df = pd.read_csv(DATA_DIR / 'new_retail_data.csv')
```

```python
print(df.shape)
df.head()
```

```python
# checking for missing vlaues
print(df.isnull().sum())
```

```python
# Handling missing values
# Drop missing values
df = df.dropna(subset=['Transaction_ID', 'Customer_ID'])
df.shape
```

```python
# fill categorical fields with placeholders (unknown)

categorical_cols = ['Gender','Customer_Segment','Shipping_Method', 'Payment_Method','Order_Status']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
#Checking the DataType
df.info()
```

```python
# converting transaction, Age and customer ID  to integer
df['Transaction_ID'] = df['Transaction_ID'].astype('Int64')
df['Customer_ID'] = df['Customer_ID'].astype('Int64')
df['Age'] = df['Age'].astype('Int64')


```

```python
# converting phone, zipcodes, and year to string
##df['Phone'] = df['Phone'].astype('str').str.replace(r'\.0$', '', regex=True)
df['Phone'] = df['Phone'].astype('str').str.replace('.0', '', regex=False)
df['Zipcode'] = df['Zipcode'].astype('str').str.replace('.0', '', regex=False)
df['Year'] = df['Year'].astype('str').str.replace('.0', '', regex=False)
```

```python
# converting date value to date type
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df.head()
```

```python
df.shape
```

```python
# drop exact duplicate
df = df.drop_duplicates()


# Drop rows with dupe transaction id
df = df.drop_duplicates(subset=['Transaction_ID'])

df.shape

```

```python
# create new columns

# Average spend per purchase
df['Avg_Spend'] = df['Total_Amount'] / df['Total_Purchases']

# Customer Age Group
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 18, 35, 45, 60, 100],
    labels=['Teen', 'Young Adults', 'Adult', 'Middle Aged', 'Senior']
)

# extract weekday from date
df['Weekday'] = df['Date'].dt.day_name()
```

```python
# a little bit numpy
# high value customer flag(1 if above median total amount. else 0)
df['High_Value'] = np.where(df['Total_Amount'] > df['Total_Amount'].median(), 1, 0)
```

```python
df.info()
```

```python
# customer rating metric

rating_matrix = df.pivot_table(index='Customer_ID',
                               columns= 'Product_Type',
                               values= 'Ratings',
                               aggfunc='mean').fillna(0)

rating_matrix
                               
```

```python
ratings_array = rating_matrix.to_numpy()
print(ratings_array)
```

```python
# customer table
customers = df[[
    "Customer_ID","Name","Email","Phone","Address","City","State","Zipcode",
    "Country","Age","Gender","Income","Customer_Segment","Age_Group","High_Value"
]].drop_duplicates(subset=["Customer_ID"]).reset_index(drop=True)

customers
```

```python
# Products table
products = df[
    [
        "Product_Category", "Product_Brand", "Product_Type", "products"
    ]
].drop_duplicates().reset_index(drop=True)
products["Product_ID"] = products.index + 1  # assign unique IDs
products = products[["Product_ID"] + [col for col in products.columns if col != "Product_ID" ]]
products
```

```python
orders = df[
    [
        "Transaction_ID", "Customer_ID", "Date", "Year", "Month", "Weekday", "Time",
        "Total_Purchases", "Total_Amount", "Shipping_Method", "products", "Order_Status"
    ]
].drop_duplicates(subset=["Transaction_ID"]).reset_index(drop=True)

orders
```

```python
payments = df[
    [
        "Transaction_ID", "Amount", "Payment_Method"
    ]
].copy().reset_index(drop=True)
payments["Payment_ID"] = payments.index + 1
payments = payments[["Payment_ID"] + [col for col in payments.columns if col != "Payment_ID"]]

payments
```

```python
# Feedback table
feedbacks = df[
    [
        "Transaction_ID", "Product_Type", "Ratings", "Feedback"
    ]
].dropna(subset=["Ratings", "Feedback"], how='all').reset_index(drop=True)  # keep only where at least 1

feedbacks["Feedback_ID"] = feedbacks.index + 1

feedbacks = feedbacks[["Feedback_ID"] + [col for col in feedbacks.columns if col != "Feedback_ID"]]

feedbacks
```

```python

# Export tab
OUT = Path(r'C:\Users\Admin\OneDrive\Documents\levelz-project\processed')
OUT.mkdir(parents=True, exist_ok=True)

customers.to_csv(OUT / 'customers.csv', index=False)
products.to_csv(OUT / 'products.csv', index=False)
orders.to_csv(OUT / 'orders.csv', index=False)
payments.to_csv(OUT / 'payments.csv', index=False)
feedbacks.to_csv(OUT / 'feedbacks.csv', index=False)
```
